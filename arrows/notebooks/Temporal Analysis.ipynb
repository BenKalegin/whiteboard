# Temporal analysis of ink drawing of arrow figures

```python
import json
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
with open('../data/arrow101.json') as f:
    curves = json.load(f)['requests'][0]['ink']
for curve in curves:
    plt.plot(curve[0], curve[1])
```

## Pen Speed

```python
plt.ylabel('speed, px/msec')    
plt.xlabel('time, ms')    

for c in range(0, len(curves)):
    x = curves[c][0]
    y = curves[c][1]
    t = curves[c][2]

    speed = [0]
    for i in range(1, len(x)):
        dx = x[i]- x[i-1]
        dy = y[i]- y[i-1]
        s = math.sqrt(dx*dx  + dy*dy) / (t[i] - t[i-1])
        speed.append(s)

    plt.plot(t, speed)
```

## Heading

```python
plt.ylabel('heading, degrees')    
plt.xlabel('time, ms')    

for c in range(0, len(curves)):
    x = curves[c][0]
    y = curves[c][1]
    t = curves[c][2]

    angle = [0]
    for i in range(1, len(x)):
        dx = x[i]- x[i-1]
        dy = y[i]- y[i-1]
        a = math.atan2(dy, dx) * 90
        angle.append(a)

    plt.scatter(t, angle)
```
