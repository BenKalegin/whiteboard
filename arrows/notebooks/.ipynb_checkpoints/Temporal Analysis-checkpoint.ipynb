# Temporal analysis of ink drawing of arrow figures

```python
import json
import numpy as np
import math
import matplotlib.pyplot as plt
from scipy import stats
from scipy import signal
from scipy.interpolate import griddata
from scipy.interpolate import interp1d
```

```python
def movingaverage(interval, window_size):
    window = np.ones(int(window_size))/float(window_size)
    return np.convolve(interval, window, 'same')    
```

```python
with open('../data/arrow101.json') as f:
    curves = json.load(f)['requests'][0]['ink']
for curve in curves:
    plt.plot(curve[0], curve[1])
```

## Pen Speed

```python
plt.ylabel('speed, px/msec')    
plt.xlabel('time, ms')    

def speed_over_n_point(x, y, t, n):
    speed = []
    for i in range(n, len(x)):
        dx = x[i]- x[i-n]
        dy = y[i]- y[i-n]
        s = math.sqrt(dx*dx  + dy*dy) / (t[i] - t[i-n])
        speed.append(s)
    return speed 

for c in range(0, len(curves)):
    x = curves[c][0]
    y = curves[c][1]
    t = curves[c][2]

n = 1   
speed = np.array(speed_over_n_point(x, y, t, n))

f = interp1d(t[n:], speed, kind="nearest")

tnew = np.linspace(t[n], t[-1], num=len(t), endpoint=True)
speednew = speed.max() - f(tnew)
plt.plot(tnew, speednew)

peak_widths = np.arange(1, len(x))
peak_indices = signal.find_peaks_cwt(speednew, peak_widths, window_size=10 )
for i in peak_indices:
    plt.axvline(tnew[i], 0)
plt.show
```

## Heading

```python
plt.ylabel('heading, degrees')    
plt.xlabel('time, ms')    

for c in range(0, len(curves)):
    x = curves[c][0]
    y = curves[c][1]
    t = curves[c][2]

    angle = [0]
    for i in range(1, len(x)):
        dx = x[i]- x[i-1]
        dy = y[i]- y[i-1]
        a = math.degrees(math.atan2(dy, dx))
        angle.append(a)

    plt.scatter(t, angle)
```

## Delta

```python
plt.ylabel('delta, ')    
plt.xlabel('time, ms')    

w = 4

for c in range(0, len(curves)):
    x = np.array(curves[c][0])
    y = np.array(curves[c][1])
    t = np.array(curves[c][2])

    angle = []    
    for i in range(0, len(x) - w):
        coefficients = np.polyfit( x[i:i+w], y[i:i+w], 1 )
        a = coefficients[0] / coefficients[1]
        #a = math.degrees(math.atan2(coefficients[0], coefficients[1]))
        angle.append(a)

    plt.plot(t[:-w], angle)
```
